```python
dfx=df({'Store':['A','B','A','C','D','A','D','A','C','B'],
        'Location':['Mumbai','Pune','Hyderabad','Mumbai','Pune','Delhi','Hyderabad','Pune','Mumbai','Pune'],
       'Sales':[40000,45000,50000,90000,89000,87000,85000,78000,89000,70000]})
dfx
```

```python
c=pd.pivot_table(dfx,'Sales','Store',aggfunc=[np.min,np.max])
d=pd.pivot_table(dfx,'Sales','Location',aggfunc=[np.min,np.max])
pd.concat((c,d),keys=['Stores','Locations'])
```
